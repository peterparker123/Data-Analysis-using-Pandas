```python
# We will choose a numeric feature to find out the mean of the Series

s = pd.read_csv("../datasets/NYC_Jobs.csv",usecols=["Salary Range From"]).squeeze()
s
```

```python
s.min()
```

```python
s = s.drop_duplicates()
s.count()
```

```python
s.rank().tolist()
```

### Series.sum([axis, skipna, numeric_only, ...])
- Return the sum of the values over the requested axis.

```python
s.sum()
```

### Series.var([axis, skipna, ddof, numeric_only]) 
- Return unbiased variance over requested axis.

```python
s.var()
```

### Series.where(cond[, other, inplace, axis, level]) 
- Replace values where the condition is False.

```python
# For e.g find the salaries where it is greater than 500
s.where(s > 85000)
```

```python
# by using other, we can replace the NaN with other value, for .e.g 0
s.where(s > 85000,other = 0)
```
