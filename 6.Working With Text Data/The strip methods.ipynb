# The strip methods

```python
import pandas as pd
df=pd.read_csv("../datasets/chicago.csv").dropna(how="all")
df.head()
```

Some times when we work with dataset, some values may have spaces at beginning or end, it is always good idea to remove
those before start working on it, for that we can use
lstrip-to remove spaces at the beginning
rstrip-to remove spaces at the end
strip- to remove spaces at both beginning and end

```python
df["Name"]=df["Name"].str.lstrip()
df.head()
```

```python
df["Position Title"]=df["Position Title"].str.rstrip()
df.head()
```

```python
df["Department"]=df["Department"].str.strip()
df.head()
```
