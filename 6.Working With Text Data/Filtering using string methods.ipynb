# Filtering using string methods

```python
import pandas as pd
df=pd.read_csv("../datasets/chicago.csv").dropna()
df.head()
```

```python
#Lets query the Position Title to contain water in it
mask=df["Position Title"].str.lower().str.contains("water")
df[mask]
```

```python
#Lets check Positional Title to start with water
mask1=df["Position Title"].str.lower().str.startswith("water")
df[mask1]
```

```python
#Lets query it for ending with ist
mask2=df["Position Title"].str.lower().str.endswith("ist")
df[mask2]
```
