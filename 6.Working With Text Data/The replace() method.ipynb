# The replace() method

```python
import pandas as pd
df=pd.read_csv("../datasets/chicago.csv").dropna(how="all")
df.head()
```

```python
df["Department"]=df["Department"].str.replace("MGMNT","MANAGEMENT")
df.head()
```

```python
df["Employee Annual Salary"]=df["Employee Annual Salary"].str.replace("$","").astype(float)
df.head()
```
