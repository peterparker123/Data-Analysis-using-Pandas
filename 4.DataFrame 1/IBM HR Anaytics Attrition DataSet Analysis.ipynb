```python
# Attrition: When an employee leaves the company due to resignation or retirement, then it is called Attrition. Employees leave the company for personal and professional reasons like retirement, lower growth potential, lower work satisfaction, lower pay rate, bad work environment, etc.

# Attrition is part and parcel of any business.

# Attrition is a cause of concern when it crosses a limit.

# This dataset has been taken from Kaggle,which will be analysed to identify various factors which contributes to employee attrition.
```

```python
path = "../datasets/"
file = os.path.join(path,'HR_Analytics.csv')
print(file)
```

```python
# Read the above csv file in the pandas dataframe
# and display the top 5 rows
df = pd.read_csv(file)
type(df)
```

```python
# display the top 5 row
df.head()
```

```python
# Adding a column to the dataframe

df["attrition"] = df["Attrition"]
df.drop(columns = ["Attrition"], inplace = True)
df.head()
```

```python
# Get the summary statistics of each attributes of the data
df.describe()
```

```python
# Calculate the attrition percentage
print("{:.2f}% of the employes resigned/retired".format(df["attrition"].value_counts()["Yes"] \
      /len(df["attrition"]) * 100))
```

```python
# Find the count of the employees who have resigned/retired
# To find the individual counts of the value, we use the value_counts() method# #
df["attrition"].value_counts()["Yes"]
```

```python
# count function
df["attrition"].count()
```

```python
import seaborn as sns
```

```python
# Plot a distribution plot or box plot, to display how many
# employees have resigned/retired 
sns.displot(data=df,x="attrition")
```

```python
# Obtain the geneder wise data for the attrition
# First, we will groupby the attirtion data based on the Gender
# Then, we will get the value_counts on the groupby data

# The groupby data returns the series
grouped_data = df.groupby("Gender")["attrition"]
print(grouped_data)
```

```python
# From the grouped_data series, return the value count
grouped_data.value_counts()
```

```python
# Check the count of the gender
["Gender"].value_counts()
```

```python
print("Total number of females who have resigned/retired, {:.2f}%".format(grouped_data.value_counts()["Female"]["Yes"]/(grouped_data.value_counts()["Male"]["Yes"] + grouped_data.value_counts()["Female"]["Yes"]) * 100))
print("Total number of males who have resigned/retired, {:.2f}%".format(grouped_data.value_counts()["Male"]["Yes"]/(grouped_data.value_counts()["Male"]["Yes"] + grouped_data.value_counts()["Female"]["Yes"]) * 100))

print("The percentage of males who have resigned/retired,{:.2f}%".format(grouped_data.value_counts()["Male"]["Yes"]/df["Gender"].value_counts()["Male"] * 100))

print("The percentage of females who have resigned/retired,{:.2f}%".format(grouped_data.value_counts()["Female"]["Yes"]/df["Gender"].value_counts()["Female"] * 100))
```

```python
# Plot the histrogram depicting the ratio of attrition amongst the genders
sns.displot(data = df, x = "Gender", hue = "attrition")
```

```python
# Attrition based on education
# Education

# 1 'Below College'
# 2 'College'
# 3 'Bachelor'
# 4 'Master'
# 5 'Doctor'

# Attrition based on Education
groupby_Education = df.groupby("attrition")["Education"]

print(100*groupby_Education.value_counts()["Yes"]/sum(groupby_Education.value_counts()["Yes"]))

print("\nOf the total employees, {:.2f}% belong to Education = 3".format(100*(groupby_Education.value_counts()[0] + groupby_Education.value_counts()[5])/sum(groupby_Education.value_counts())))
print("Of the total employees resigned/retired, {:.2f}% were with Education = 3".format(groupby_Education.value_counts()["Yes"][3]/sum(groupby_Education.value_counts()["Yes"])*100))


sns.displot(data = df, x = "Education", hue = "attrition")
sns.displot(data = df, x = "Education", hue = "attrition", col = "Gender")
sns.displot(data = df, x = "Education", hue = "attrition", col = "MaritalStatus")
```
